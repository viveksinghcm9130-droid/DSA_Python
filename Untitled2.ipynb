# Selection sort

```python
def selection(arr):
    n=len(arr)
    for i in range(n-1):
        position=i
        for j in range(i+1,n):
            if arr[j]<arr[position]:
                position=j
        temp=arr[position]
        arr[position]=arr[i]
        arr[i]=temp
    return arr
```

```python
a=[4,3,2,5,55,2,1,33,4,6,7,4]
print(selection(a))
```

# Insertion sort

```python
def insertion_sort(arr):
    n=len(arr)
    for i in range(1,n):
        cvalue=arr[i]
        position=i
        while position>0 and arr[position-1]>cvalue:
            arr[position]=arr[position-1]
            position-=1
        arr[position]=cvalue
    return arr
            
            
            
```

```python
a=[4,3,2,5,55,2,1,33,4,6,7,4]
print(insertion_sort(a))
```

# Bubble sort

```python
def Bubble_sort(arr):
    n=len(arr)
    for i in range(n-1,0,-1):
        j=0
        while j<i:
            if arr[j]>arr[j+1]:
                arr[j+1],arr[j]=arr[j],arr[j+1]
            j+=1
    return arr
            
        
```

```python
a=[4,3,2,5,55,2,1,33,4,6,7,4]
print(Bubble_sort(a))
```
