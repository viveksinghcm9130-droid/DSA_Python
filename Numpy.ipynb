```python
def fib(n):
    if n<1:
        return 1
    return fib(n-1)+fib(n-2)

print(fib(4))
```

```python
def factorial(n):
    if n<1:
        return 1
    return factorial(n-1)*n

print(factorial(5))
```

```python
def recurrsion(arr,key,l,r):
    if l>r:
        return -1
    else:
        mid=(l+r)//2
        if arr[mid]>key:
            recurrsion(arr,key,0,mid-1)
        elif arr[mid]<key:
            recurrsion(arr,key,mid+1,r)
        elif arr[mid]==key:
            print(f"{key} found")
            return
    
        
```

```python
l1=[1,2,3,5,7,8,10,22,44,55]
recurrsion(l1,10,0,len(l1))
```

```python
import itertools
a = [2, 4, 6, 7, 8, 10, 11, 12, 13]

# Grouping consecutive even and odd numbers
g_data = itertools.groupby(a, key=lambda x: 'Even' if x % 2 == 0 else 'Odd')
print(g_data)
```
